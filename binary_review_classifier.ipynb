```python
import pandas as pd
#import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
df = pd.read_csv('parsed_reviews3.csv')
```

```python
#–í—ã–±–æ—Ä —Ç–æ–ª—å–∫–æ –Ω—É–∂–Ω–æ–π –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –∏–∑ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞
def add_space(text):
    return text + ' '


df['title'] = df['title'].apply(add_space)
df['body'] = df['title'] + df['body']
df = df[['body','rating']]

#0 –¥–ª—è –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã—Ö –æ—Ç–∑—ã–≤–æ–≤ –∏ 1 –¥–ª—è –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö
df.loc[df['rating'] < 3, 'rating'] = 0
df.loc[df['rating'] >= 3, 'rating'] = 1
```

```python
#–û—á–∏—Å—Ç–∫–∞ —Ç–µ–∫—Å—Ç–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö
def remove_trash(text):
    trash = ['!','.',',','?','\xa0','&lt','üòÇ','‚ù§Ô∏è','üëç','p&gt','~','\r','&quot','\n',';','/','br&gt','*','$','+','\t','&gt',
             'li&gt','ul&gt','&#039','ü•≥',':','=','(',')','-']
    for sym in trash:
        text = text.replace(sym, ' ')
        
    while '  ' in text:
        text = text.replace('  ', ' ')
        
    text = text.lower()
    
    return text
 


df['body'] = df['body'].apply(remove_trash)
```

```python
df
```

```python
#–ü–æ–ª—É—á–µ–Ω–∏–µ —Å–ø–∏—Å–∫–æ–≤ –¥–ª—è ML –º–æ–¥–µ–ª–∏ –∏–∑ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞
body = df['body'].tolist()
rating = df['rating'].tolist()
```

```python
body
```

```python
#–í–µ–∫—Ç–æ—Ä–∏–∑–∞—Ü–∏—è
count_vect = CountVectorizer()
body_vect = count_vect.fit_transform(body)
body_vect.shape
```

```python
#–†–∞–∑–¥–µ–ª–µ–Ω–∏–µ –Ω–∞ –æ–±—É—á–∞—é—â—É—é –∏ —Ç–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫–∏
x_train, x_test, y_train, y_test = train_test_split(body_vect, rating, random_state=0)
```

```python
#–û–±—É—á–µ–Ω–∏–µ
clf = MultinomialNB().fit(x_train, y_train)
```

```python
#–ü–æ–¥—Å—á—ë—Ç —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏ –º–æ–¥–µ–ª–∏
clf.score(x_test,y_test)
```
